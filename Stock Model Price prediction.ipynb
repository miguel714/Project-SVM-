```python
# Import the Libraries 

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
import pandas_datareader.data as web
import datetime as dt
```

```python
# Read the data

dataset = pd.read_csv('istanbul_stock_exchange.csv')

dataset 
```

```python
# This is to return data of the last rows since it's the latest stock prices

dataset.tail()
```

```python
# Chose BOVESPA stock since it's the same one we chose for our LSTM model in order to compare and contrast between the two

df = dataset[['BOVESPA']]

# Printed out the latest stock prices from BOVESPA

dataset[['BOVESPA']].tail()
```

```python
# Created a prediction column that moved up to 10 days

df['Prediction'] = df[['BOVESPA']].shift(-10)

print(df)
```

```python
# Created an X variable in order set it as a dataset and then turned it into a nump array

X = np.array(df.drop(['Prediction'], 1))

# Remove the last 10 rows

X = X[:-10]

print(X)
```

```python
# Created a Y variable, and did the same with the X variable, setting it as a dataset and then turning it into a numpy array

y = np.array(df['Prediction'])

# Remove Last 10 rows

y = y[:-10]

print(y)
```

```python
# We then split the dataset into a train and test with 20% being tested and 80% being trained

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# We create the SVM Model

svr = SVR(kernel= 'rbf', C=100, gamma=0.5)

# We then train the model 

svr.fit(x_train, y_train)
```

```python
# The output below is the confidence value score

SVM_confidence = svr.score(x_test, y_test)

print('SVM confidence: ', SVM_confidence)
```

```python
# The output below are the latest stock prices within the 10 days

Stock_forecasted = np.array(df.drop(['Prediction'],1))[-10:]

print(Stock_forecasted)
```

```python
#  The output below are the predicted stock prices for the next 10 days

SVM_prediction = svr.predict(Stock_forecasted)

print(SVM_prediction)
```
